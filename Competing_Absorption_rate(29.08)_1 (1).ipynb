```python
import numpy as np
import matplotlib.pyplot as plt


def Competetive_absorbtion_per_molecule_per_second(wavelength, Inc_Area, Intensity, Path_length, Conc_A, Conc_B, epsilon_A, epsilon_B):
    """
    Calculate the transmitted intensity of light through a medium with two competing absorbing species.

    Parameters:
    wavelength (float): Wavelengths of light (in nm).
    Inc_Area (float): Incident area of the light beam (in cm^2).
    Intensity (float): Incident light intensity at each wavelength (in mW/cm^2).
    Path_length (float): Path length through the medium (in cm).
    Conc_A (float): Concentration of species A (in mM).
    Conc_B (float): Concentration of species B (in mM).
    epsilon_A (int): Molar absorptivity of species A at each wavelength (in mM^-1 cm^-1).
    epsilon_B (int): Molar absorptivity of species B at each wavelength (in mM^-1 cm^-1).
    """

    #Define constants

    h = 6.62607015e-34  # in JÂ·s
    c = 3e8  # in m/s
    avogadro_number = 6.02214076e23  # in mol^-1

    # Calculate the absorbance for each species
    Abs_A = epsilon_A * Conc_A * Path_length
    Abs_B = epsilon_B * Conc_B * Path_length

    # Total absorbance is the sum of individual absorbances
    Total_Abs = Abs_A + Abs_B

    # Calculate transmitted intensity using Beer-Lambert law 
    # from there calculate the amount of absorbed phtons by each species

    Transmitted_Intensity = Intensity * 10**(-Total_Abs)
    Fraction_Absorbed = 1 - (Transmitted_Intensity / Intensity)
    Photon_flux = (wavelength*Intensity*Inc_Area)/(h*c)
    photons_absorbed = Fraction_Absorbed*Photon_flux
    Fractional_A = Abs_A/Total_Abs
    Fractional_B = Abs_B/Total_Abs

    # Calculate the excitation rate per molecule

    Excitation_per_molecule_A = (1/(Inc_Area*Path_length*(1/1000)*avogadro_number*Conc_A))*photons_absorbed*Fractional_A
    Excitation_per_molecule_B = (1/(Inc_Area*Path_length*(1/1000)*avogadro_number*Conc_B))*photons_absorbed*Fractional_B

    return Excitation_per_molecule_A, Excitation_per_molecule_B
```

```python
wavelength = 0.0000005
Incident_area = 1
Intensity = 0.1
Path_length = 1
Epsilon_A = 10000
Epsilon_B = 7000
Conc_B = 0.00001
Conc_A = np.linspace(Conc_B, 10*Conc_B, 20)
Exc_A=[]
Exc_B=[]

Exc_A, Exc_B = Competetive_absorbtion_per_molecule_per_second(wavelength, Incident_area, Intensity, Path_length, Conc_A, Conc_B, Epsilon_A, Epsilon_B)


plt.figure()
plt.xlabel('Concentration')
plt.ylabel('Excitation per molecule per second')
plt.title('Excitation rate for each especies as concentration of A varies')
plt.plot(Conc_A, Exc_A, label='Excitation rate per A molecule')
plt.plot(Conc_A, Exc_B, label='Excitation rate per B molecule')
plt.legend(loc='best')

```
